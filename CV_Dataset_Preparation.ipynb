```python
import numpy as np
import random
import cv2

def sp_noise(image,prob):
    '''
    Add salt and pepper noise to image
    prob: Probability of the noise
    '''
    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output
```

```python
import cv2
from google.colab.patches import cv2_imshow

for x in range (1, 792):
  img = cv2.imread('1 (' + str(x) + ').jpg', cv2.IMREAD_UNCHANGED)
  
  print('Original Dimensions : ',img.shape)
  
  width = 512
  height = 512
  dim = (width, height)
  
  # resize image
  resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
  gray_resized = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
  gray2rgb = cv2.cvtColor(gray_resized,cv2.COLOR_GRAY2RGB)
  noise_img = sp_noise(gray2rgb,0.05)
  #print('Resized Dimensions : ',resized.shape)
  print('Grayscale Dimensions : ',noise_img.shape)
  #cv2_imshow(resized)
  cv2.imwrite('/content/gdrive/MyDrive/CVProject/Dataset/color/'+ str(x) +'.jpg', resized)
  cv2.imwrite('/content/gdrive/MyDrive/CVProject/Dataset/grayscale/'+ str(x) +'.jpg', noise_img)
```
